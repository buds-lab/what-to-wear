```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load file
df_references = pd.read_csv("../data/references.csv")
```

```python
df_references.head()
```

```python
def plotHistogram(df, column):
    plt.figure()
    ax = df[column].value_counts().plot(kind = 'bar', figsize = (12,10))
    ax.set_title(column, fontsize=18)
    return
```

# Histograms and Figures

```python
print("Total number of papers: {}".format(df_references.shape[0]))

# papers by theme
plotHistogram(df_references, "Theme")
# papers by year
plotHistogram(df_references, "Date")
# papers by users
plotHistogram(df_references, "Targeted user")
# papers by wearable type
plotHistogram(df_references, "Wearable type")
# papers by objective
plotHistogram(df_references, "Objectives")
```
